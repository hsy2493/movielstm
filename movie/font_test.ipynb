```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os
import numpy as np

# 1. 현재 작업 디렉토리 가져오기
current_dir = os.getcwd()

# 2. 폰트 파일 경로 설정 (상대 경로)
# font_path = os.path.join(current_dir, 'font', 'NanumGothic.ttf')  # 폰트 파일명을 실제 파일명으로 변경
font_path = '/System/Library/Fonts/Supplemental/AppleGothic.ttf'

# 3. 폰트 파일 존재 확인
if not os.path.exists(font_path):
    print(f"오류: 폰트 파일 '{font_path}'을 찾을 수 없습니다.")
    print("주피터 노트북 파일과 폰트 폴더가 같은 디렉토리에 있는지 확인하세요.")
    exit()

# 4. FontProperties 객체 생성 및 폰트 이름 얻기
try:
    font_name = fm.FontProperties(fname=font_path).get_name()
    plt.rcParams['font.family'] = font_name
except Exception as e:
    print(f"오류: 폰트 로딩 실패: {e}")
    print("폰트 경로 및 파일명을 확인하고, 폰트가 손상되지 않았는지 확인하세요.")
    exit()

# 5. 한글이 포함된 데이터 생성 (시각화용)
data = {
    '제품': ['사과', '바나나', '오렌지', '딸기', '포도'],
    '판매량': [100, 150, 80, 120, 90]
}

# 6. 막대 그래프 그리기
try:
    plt.figure(figsize=(8, 6))  # 그래프 크기 설정 (선택 사항)
    plt.bar(data['제품'], data['판매량'])
    plt.xlabel('제품', fontsize=12)
    plt.ylabel('판매량', fontsize=12)
    plt.title('과일별 판매량', fontsize=14)
    plt.xticks(rotation=45, ha='right')  # x축 레이블 회전 (선택 사항)
    plt.tight_layout()  # 레이아웃 조정 (레이블 겹침 방지)
    plt.show()

except Exception as e:
    print(f"오류: 그래프 그리기 실패: {e}")
    print("폰트 설정이 올바른지, Matplotlib 버전이 최신인지 확인하세요.")
```
